#which one should be used(pyplot or OO plot)
when ploting something quickly we use pyplot method
when ploting something advanced use OO plot method

```python
heart_disease
```

```python
over_50=heart_disease[heart_disease["age"]>50]
```

```python
over_50
```

```python
#pyplot method
over_50.plot(x='chol',y='age',c='target',kind="scatter")
```

```python
#OO method object oriented method
fig,ax=plt.subplots(figsize=(10,6))
#over_50.plot(x='age',y='chol',c='target',kind='scatter',ax=ax)
over_50.plot(x="age",y="chol",c="target",kind="scatter",ax=ax)
#ax.set_xlim([45,100]) for axis limit size
```

```python
#oo method from scratch
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])
#customize plot
ax.set(title="heart disease and cholesteorl",xlabel="age",ylabel="chol");
#add a legend
ax.legend(*scatter.legend_elements(),title="target")
#add a horizontal line
ax.axhline(over_50["chol"].mean())
#for dash line
ax.axhline(over_50["chol"].mean(),linestyle="--")
```
